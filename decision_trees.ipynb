## Student Dataset

```python
import numpy as np
import pandas as pd
```

The 'ClassificationDataSetforStudents.xls' file was converted to two csv files: 'traindata.csv' and 'testdata.csv'.

```python
train = pd.read_csv('traindata.csv', header=None)
test = pd.read_csv('testdata.csv', header=None)
```

```python
train_X = train.drop(train.columns[23], axis=1)
train_y = train.iloc[:,23]
```

## Decision Tree Classifier

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import StratifiedKFold
from sklearn.model_selection import cross_val_score
from sklearn import tree
from matplotlib import pyplot as plt
```

```python
clf = DecisionTreeClassifier(random_state=0)
clf.fit(train_X, train_y)
```

```python
metrics = ['precision', 'recall', 'accuracy']
for m in metrics:
    scores = cross_val_score(clf, train_X, train_y, scoring=m, cv=10)
    print(m, ": ", scores)
    print("average", m, ": ", scores.mean())
```

## AdaBoost

```python
from sklearn.ensemble import AdaBoostClassifier
```

```python
boosted_clf = AdaBoostClassifier()
boosted_clf.fit(train_X, train_y)
```

```python
metrics = ['precision', 'recall', 'accuracy']
for m in metrics:
    scores = cross_val_score(boosted_clf, train_X, train_y, scoring=m, cv=10)
    print(m, ": ", scores)
    print("average", m, ": ", scores.mean())
```

```python
# different learning rates
lr = [0.001, 0.01, 0.1]
for r in lr:
    boosted_clf = AdaBoostClassifier(learning_rate=r)
    print("=================================================")
    print("learning rate", r)
    for m in metrics:
        scores = cross_val_score(boosted_clf, train_X, train_y, scoring=m, cv=10)
        print(m, ": ", scores)
        print("average", m, ": ", scores.mean())
```

Changes in performance after using AdaBoost, with best accuracy obtained using learning rate 0.01:

```python
print((0.7035299677622854-0.393024995552109)/0.393024995552109*100, "percent increase in precision")
print((0.33391880018179065-0.42044114528101806)/0.42044114528101806*100, "percent increase in recall")
print((0.8217307692307692-0.7277307692307692)/0.7277307692307692*100, "percent increase in accuracy")
```

## Gradient Boosting Classifier

```python
from sklearn.ensemble import HistGradientBoostingClassifier
```

```python
hgbc = HistGradientBoostingClassifier(categorical_features=[0,1,2], learning_rate=0.05)
hgbc.fit(train_X, train_y)
scores = cross_val_score(hgbc, train_X, train_y, scoring='accuracy', cv=10)
print("accuracy:", scores)
print("average accuracy:", scores.mean())
```
